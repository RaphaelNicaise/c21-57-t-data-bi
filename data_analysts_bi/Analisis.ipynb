```python
pais_agrupado_para_tasa
```

```python
pais_agrupado_para_tasa['tasa_abandono'] = pais_agrupado_para_tasa ['tasa_abandono']
print(pais_agrupado_para_tasa[['Country', 'tasa_abandono']])
```
